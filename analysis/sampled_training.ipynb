```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
neurognn_detection_auroc = {
    0.2: 0.712,
    0.4: 0.760,
    0.6: 0.792,
    0.8: 0.832,
    1: 0.850
}

dcrnn_detection_auroc = {
    0.2: 0.707,
    0.4: 0.711,
    0.6: 0.731,
    0.8: 0.826,
    1: 0.820
}

neurognn_classification_f1 = {
    0.2: 0.792,
    0.4: 0.802,
    0.6: 0.827,
    0.8: 0.817,
    1: 0.790
}

dcrnn_classification_f1 = {
    0.2: 0.606,
    0.4: 0.689,
    0.6: 0.679,
    0.8: 0.704,
    1: 0.701
}
```

```python
# Plot for AUROC
plt.figure(figsize=(10, 6))
plt.plot(list(neurognn_detection_auroc.keys()), list(neurognn_detection_auroc.values()), marker='o', label='NeuroGNN', color='blue')
plt.plot(list(dcrnn_detection_auroc.keys()), list(dcrnn_detection_auroc.values()), marker='^', label='DCRNN', color='red')
plt.title('AUROC for Seizure Detection')
plt.xlabel('Training Data Ratio')
plt.ylabel('AUROC')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig('figures/sampled-train/sampled_seizure_detection_auroc.png', dpi=300, bbox_inches='tight', transparent=True)
```

```python
# Plot for F1
plt.figure(figsize=(10, 6))
plt.plot(list(neurognn_classification_f1.keys()), list(neurognn_classification_f1.values()), marker='o', label='NeuroGNN', color='blue')
plt.plot(list(dcrnn_classification_f1.keys()), list(dcrnn_classification_f1.values()), marker='^', label='DCRNN', color='red')
plt.title('Weighted F1 for Seizure Classification')
plt.xlabel('Training Data Ratio')
plt.ylabel('Weighted F1 Score')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig('figures/sampled-train/sampled_seizure_classification_f1.png', dpi=300, bbox_inches='tight', transparent=True)
```

### Plots comparing the performance ratio for using different sample portions to using all data 

```python
# Calculate performance as a percentage compared to using all training samples
def calculate_percentage(dictionary):
    max_value = dictionary[1]  # this is the value at key=1
    return {k: (v / max_value) * 100 for k, v in dictionary.items()}

neurognn_detection_auroc_percentage = calculate_percentage(neurognn_detection_auroc)
dcrnn_detection_auroc_percentage = calculate_percentage(dcrnn_detection_auroc)

neurognn_classification_f1_percentage = calculate_percentage(neurognn_classification_f1)
dcrnn_classification_f1_percentage = calculate_percentage(dcrnn_classification_f1)
```

```python
# Plot for AUROC percentage
plt.figure(figsize=(10, 6))
plt.plot(list(neurognn_detection_auroc_percentage.keys()), list(neurognn_detection_auroc_percentage.values()), marker='o', label='NeuroGNN', color='blue')
plt.plot(list(dcrnn_detection_auroc_percentage.keys()), list(dcrnn_detection_auroc_percentage.values()), marker='^', label='DCRNN', color='red')
plt.title('AUROC Percentage Change for Seizure Detection Compared to Training on Full Data')
plt.xlabel('Training Data Ratio')
plt.ylabel('AUROC (%) Change')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig('figures/sampled-train/sampled_seizure_detection_auroc_percentage.png', dpi=300, bbox_inches='tight', transparent=True)
```

```python
# Plot for F1 percentage
plt.figure(figsize=(10, 6))
plt.plot(list(neurognn_classification_f1_percentage.keys()), list(neurognn_classification_f1_percentage.values()), marker='o', label='NeuroGNN', color='blue')
plt.plot(list(dcrnn_classification_f1_percentage.keys()), list(dcrnn_classification_f1_percentage.values()), marker='^', label='DCRNN', color='red')
plt.title('Weighted F1 Percentage Change for Seizure Classification Compared to Training on Full Data')
plt.xlabel('Training Data Ratio')
plt.ylabel('Weighted F1 Score (%) Change')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig('figures/sampled-train/sampled_seizure_classification_f1_percentage.png', dpi=300, bbox_inches='tight', transparent=True)
```
