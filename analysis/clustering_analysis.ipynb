```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Random seed for reproducibility
np.random.seed(0)

# Generate synthetic data for Method A (tight clusters)
centers_A = [(2, 2), (8, 8), (2, 8), (8, 2)]
X_A, y_A = make_blobs(n_samples=300, centers=centers_A, cluster_std=0.5, shuffle=True)

# Generate synthetic data for Method B (more spread-out clusters)
centers_B = [(2, 2), (8, 8), (2, 8), (8, 2)]
X_B, y_B = make_blobs(n_samples=300, centers=centers_B, cluster_std=1.3, shuffle=True)

# Plot the embeddings
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plotting for Method A
ax1.scatter(X_A[y_A == 0][:, 0], X_A[y_A == 0][:, 1], c='red', label='Class 0', s=50)
ax1.scatter(X_A[y_A == 1][:, 0], X_A[y_A == 1][:, 1], c='blue', label='Class 1', s=50)
ax1.scatter(X_A[y_A == 2][:, 0], X_A[y_A == 2][:, 1], c='green', label='Class 2', s=50)
ax1.scatter(X_A[y_A == 3][:, 0], X_A[y_A == 3][:, 1], c='yellow', label='Class 3', s=50)
ax1.set_title("Method A")
ax1.legend()

# Plotting for Method B
ax2.scatter(X_B[y_B == 0][:, 0], X_B[y_B == 0][:, 1], c='red', label='Class 0', s=50)
ax2.scatter(X_B[y_B == 1][:, 0], X_B[y_B == 1][:, 1], c='blue', label='Class 1', s=50)
ax2.scatter(X_B[y_B == 2][:, 0], X_B[y_B == 2][:, 1], c='green', label='Class 2', s=50)
ax2.scatter(X_B[y_B == 3][:, 0], X_B[y_B == 3][:, 1], c='yellow', label='Class 3', s=50)
ax2.set_title("Method B:")
ax2.legend()

plt.tight_layout()
```
